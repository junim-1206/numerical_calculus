# Zero da Função por meio do método da bissecção

#### 1.Definição da função e valores x e y

```python
# Importando bibliotecas necessárias
import numpy as np
import matplotlib.pyplot as plt

#Função proposta pelo problema 1
def f(x):
    return ((np.pi * x ** 2) * (3 * 1 - x))/3 - 0.5

#Função dentro dos valores dos parâmetros
def fun(xi, xs, precisao, fun):
    x = np.linspace(xi, xs, int(abs(xs-xi)/ precisao))
    y = fun(x)
    return x, y

#Plotagem dos valores delimitados
x, y = fun(0, 1, 10e-3, f)

plt.plot(x, y, '.b')
plt.xlabel('Altura')
plt.ylabel('Função da Altura')
plt.show()
```

#### 2.Calculando valor da raíz pela bissecção dos valores superiores e inferiores de função

```python
#Função geral do cálculo do método da bissecção
def calcula_bis(xi, xs, precisao, imax, funcao):
    lista_xr = []
    for i in range(imax):
        fi = funcao(xi)
        fs = funcao(xs)
        xr = (xi + xs) / 2
        bis = funcao(xr)
        lista_xr.append(xr)
        delta = abs(xs - xi) / 2

        if fi * fs > 0:
            print('Infelizmente não há raíz para a função nesse intervalo de valores, valores padrão:')
            break

        if delta < precisao and abs(bis) < precisao:
            break
        
        if bis * fs > 0:
            xs = xr
            fs = bis
        else:
            xi = xr
            fi = bis

    fig, ax = plt.subplots()
    ax.plot(range(len(lista_xr)), lista_xr, 'b')
    ax.plot(range(len(lista_xr)), lista_xr, '.k')
    ax.set_xlabel('Iteração (#)')
    ax.set_ylabel('Raíz da função')
    plt.grid()
    return print(f'O valor encontrado na iteração {i} para uma precisão {precisao} é de {xr}'), ax
```

#### 3. Resolução do Problema 1

```python
calcula_bis(0.25, 0.5, 10e-5, 20, f)
```

```python
calcula_bis(2.5, 3, 10e-5, 20, f)
```

#### 4.Resolução do Problema 2

##### 4a) Plotagem dos gráficos

```python
#Função dada pelo problema 2
def f(x):
    return (2.5 / (120 * 50000 * 30000 * 600)) * (-x ** 5 + 2 * 600 ** 2 * x ** 3 - 600 ** 4 * x)

#Derivada da função do problema 2, calculada e aplicada ao problema
def df(x):
    return ((-5 * x ** 4 * 2.5) + (6 * 600 ** 2 * x ** 2 * 2.5) - (600 ** 4 * 2.5)) / (120 * 50000 * 30000 * 600)

#Criando valores para o x do problema
x = np.linspace(0, 600, num=100)

#Dividindo em dois plots e fazendo a plotagem
figure, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].plot(x, f(x))
ax[0].set_title('Deformação')
ax[0].set_xlabel('Distância (cm)')
ax[0].set_ylabel('Deformação (cm)')

ax[1].plot(x, df(x))
ax[1].set_title('Derivada da deformação')
ax[1].set_xlabel('Distância (cm)')
ax[1].set_ylabel('Derivada')
plt.tight_layout()
plt.show()
```

##### Encontrando valor da raíz nula

```python
#IMPORTANTE: Ao tentar usar os valores de xi e xs como 0 e 600 o valor do xr acabou convergindo para um valor que NÃO a raíz da função
calcula_bis(200, 300, 10e-8, 50, df)
```

#### 5.Resolução do Problema 3

```python
#Definindo a função considerando os valores dados
def f(x):
    return 5 - 20 * (np.exp(-0.2 * x) - np.exp(-0.75 * x))


#Resolvendo o problema usando a função, usando valores de precisão e iteração mais altos para melhor convergência do valor
calcula_bis(0, 2, 10e-4, 100, f)
```

#### 6.Resolução do Problema 4

```python
#Definindo função por meio dos dados do problema
def y(ta):
    return (ta / 12) * np.cosh((12 / ta) * 50) + 6 - (ta / 12) - 15

#Visualisando o problema
x = np.linspace(-2, 2, num=600)
plt.plot(x, y(x))
plt.show()
```
